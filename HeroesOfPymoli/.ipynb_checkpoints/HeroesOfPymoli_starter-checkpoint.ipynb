```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
IDS = purchase_data["SN"].value_counts()
TotalPlayers = len(IDS)
pd.DataFrame({"Total Players": [TotalPlayers]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Average_Price = purchase_data["Price"].mean()
Unique_Items = len(purchase_data["Item ID"].value_counts())
Number_of_Purchases = purchase_data["Item Name"].count()
Total_Revenue = purchase_data["Price"].sum()
Purchasing_Analysis = pd.DataFrame({"Number of Unique Items": [Unique_Items], 
                                    "Average Price": [Average_Price],
                                   "Number of Purchases": [Number_of_Purchases],
                                   "Revenue": [Total_Revenue]})                                                     
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:,.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Revenue"].map("${:,.2f}".format)
Purchasing_Analysis 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
PlayerDemographic = purchase_data.loc[:,["Gender", "SN", "Age"]]
PlayerDemographic = PlayerDemographic.drop_duplicates()
TotalPlayers = PlayerDemographic.count()[0]
#TotalPlayers 
PlayerDemographicTotals = PlayerDemographic["Gender"].value_counts()
#PlayerDemograhicTotals
PlayerDemographicPercent = PlayerDemographicTotals / TotalPlayers 
#PlayerDemographicPercent
GenderDemographics = pd.DataFrame({"Total Counts": PlayerDemographicTotals, "Percentage of Players": PlayerDemographicPercent})
GenderDemographics["Percentage of Players"] = GenderDemographics["Percentage of Players"].map("{:,.2%}".format)
GenderDemographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Gender_Purchase_Price_Total = purchase_data.groupby(["Gender"]).sum()["Price"]
Gender_Purchase_Price_Average = purchase_data.groupby(["Gender"]).mean()["Price"]
Gender_Purchase_Count = purchase_data.groupby(["Gender"]).count()["Price"]
Gender_Purchase_Per_Person = Gender_Purchase_Price_Total / GenderDemographics["Total Counts"] 

Gender_Purchase_df = pd.DataFrame({"Purchase Count": Gender_Purchase_Count, 
                                           "Total Purchase Value": Gender_Purchase_Price_Total,
                                           "Average Purchase Price": Gender_Purchase_Price_Average,
                                           "Average Purchase Total Per Person": Gender_Purchase_Per_Person})

Gender_Purchase_df["Total Purchase Value"] = Gender_Purchase_df["Total Purchase Value"].map("${:,.2f}".format)
Gender_Purchase_df["Average Purchase Price"] = Gender_Purchase_df["Average Purchase Price"].map("${:,.2f}".format)
Gender_Purchase_df["Average Purchase Total Per Person"] = Gender_Purchase_df["Average Purchase Total Per Person"].map("${:,.2f}".format)
Gender_Purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_demographics_df = pd.DataFrame({"Age": PlayerDemographic["Age"], "SN": PlayerDemographic["SN"]})

bin_labels = ["< 10", "10-4", "15-9", "20-24", "25-29", "30-34", "35-39", "40 >"]
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 999.99]

PlayerDemographic["Age Ranges"] = pd.cut(PlayerDemographic["Age"], bins, labels=bin_labels)

age_demographic_totals = PlayerDemographic["Age Ranges"].value_counts()
age_demographic_percent = age_demographic_totals / TotalPlayers
age_demographic_percent

age_demographics_df = pd.DataFrame({"Total Counts": age_demographic_totals, "Percentage of Players": age_demographic_percent})
age_demographics_df = age_demographics_df.sort_index()
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:,.2%}".format)
age_demographics_df
```
